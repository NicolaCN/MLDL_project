```python
def updateP(p,k):
    for i in range(p.size):
        if p[i]==0:
            p[i]=np.random.binomial(1,k,1)
        else:
            p[i]=np.random.binomial(1,1-k,1)
```

```python
k=1/4
p=np.random.binomial(1,k,n)
print(p)
for i in range(100):
    updateP(p,k)
    print(p)
```
